Exploratory Data Analysis

```python
df.columns
```

```python
sns.countplot(x= 'Gender', data=df)
```

```python
ax = sns.countplot(x ='Gender', data =df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen =df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x= 'Gender', y='Amount', data= sales_gen)    
```

```python
df.columns
```

```python
sns.countplot(data=df, x= 'Age Group', hue = 'Gender')
# hue is useful for Division 
```

```python
ax = sns.countplot(data=df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount Vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y = 'Amount', data = sales_age)
```

```python
#State Vs Orders
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.barplot(x='State', y = 'Orders', data = sales_state)
```

```python
#State Vs Orders (To the X values names we can use)
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State', y = 'Orders', data = sales_state)
```

```python
#Sales amount from top 10 states
sales_smount_by_states = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State', y = 'Amount', data = sales_smount_by_states)
```

```python
#Marital_Status
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(5,2)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Marital_Status with Gender
ax = sns.countplot(data = df, x = 'Marital_Status', hue = 'Gender')
sns.set(rc={'figure.figsize':(5,2)})
for bars in ax.containers:
    ax.bar_label(bars)
```

Product Category


```python
df.columns
```

```python
Sales_Product = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=Sales_Product, x = 'Product_Category', y = 'Amount')
```

Products VS Orders

```python
Product_Orders = df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=Product_Orders, x = 'Product_Category', y = 'Orders')

sns.set(rc={'figure.figsize':(20,5)})
```
